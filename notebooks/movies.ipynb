```python
df.belongs_to_collection.fillna('{\'name\':\'sin datos\' }', inplace=True)
df.belongs_to_collection
```

```python
import ast
dicc = ast.literal_eval(df.belongs_to_collection[0])
dicc['name']
```

```python
# Supongamos que tienes un DataFrame llamado 'df' con una columna 'belongs_to_collection'

# Función para extraer el valor 'name' de cada elemento en la columna 'belongs_to_collection'
def extract_name(row):
    try:
        dicc = ast.literal_eval(row)
        return dicc['name']
    except (ValueError, SyntaxError, KeyError):
        return None
    except TypeError:
        return None

# Crear la nueva columna 'name_collection'
df['name_collection'] = df['belongs_to_collection'].apply(extract_name)
```

```python
df['genres'] = df['genres'].apply(lambda x: ast.literal_eval(x) if pd.notnull(x) else [])
df['genres'] = df['genres'].apply(lambda x: [valor['name'] for valor in x] if isinstance(x, list) else x)
```

```python
df['production_companies'] = df['production_companies'].apply(lambda x: ast.literal_eval(x) if pd.notnull(x) else [])
df['production_companies'] = df['production_companies'].apply(lambda x: [valor['name'] for valor in x] if isinstance(x, list) else x)
```

```python
df['production_countries'] = df['production_countries'].apply(lambda x: ast.literal_eval(x) if pd.notnull(x) else [])
df['production_countries'] = df['production_countries'].apply(lambda x: [valor['name'] for valor in x] if isinstance(x, list) else x)
```

```python
df['spoken_languages'] = df['spoken_languages'].apply(lambda x: ast.literal_eval(x) if pd.notnull(x) else [])
df['spoken_languages'] = df['spoken_languages'].apply(lambda x: [valor['name'] for valor in x] if isinstance(x, list) else x)
```
